# Haiti Crop Productivity Analysis

This notebook uses Enhanced Vegetation Index (EVI) from MODIS as a proxy for crop yield. EVI measures canopy greeness by analyzing spectral data through satellite imagery. EVI serves as a proxy for crop yield because it measures photosynthetic activity and green biomass, which strongly correlate with a crop's potential productivity and final harvest output. 

## Methodology Summary

1. Download imagery from MODIS for Haiti from 2012-2024. 
2. Get Crop Mask data for Haiti from ESA
3. Estimate area of crop land in each admin region 
4. Use 2021-2023 data as a baseline to identify growing season in Haiti
5. Use the growing season to track difference in EVI in 2024 compared to a 10 year average (2012-2022)
6. Plot trends in EVI over the last decade

```python
# MODIS Data for EVI
start_period = ee.Date("2012-01-01")
end_period = ee.Date("2024-12-31")

terra = (
    ee.ImageCollection("MODIS/061/MOD13Q1")
    .select(["EVI", "SummaryQA", "DetailedQA"])
    .filterDate(start_period, end_period)
)
aqua = (
    ee.ImageCollection("MODIS/061/MYD13Q1")
    .select(["EVI", "SummaryQA", "DetailedQA"])
    .filterDate(start_period, end_period)
)

modis = terra.select("EVI").merge(aqua.select("EVI"))
modis = modis.sort("system:time_start")

#modis
#----------------------BITWISE MASKING---------------------------------------
def bitwiseExtract(value, fromBit, toBit=None):
    if toBit == None:
        toBit = fromBit
    maskSize = ee.Number(1).add(toBit).subtract(fromBit)
    mask = ee.Number(1).leftShift(maskSize).subtract(1)
    return value.rightShift(fromBit).bitwiseAnd(mask)


# Applying the SummaryQA  and DetailedQA
def modisQA_mask(image):
    sqa = image.select("SummaryQA")
    dqa = image.select("DetailedQA")
    viQualityFlagsS = bitwiseExtract(sqa, 0, 1)
    viQualityFlagsD = bitwiseExtract(dqa, 0, 1)
    # viUsefulnessFlagsD = bitwiseExtract(dqa, 2, 5)
    viSnowIceFlagsD = bitwiseExtract(dqa, 14)
    viShadowFlagsD = bitwiseExtract(dqa, 15)
    # Good data, use with confidence
    mask = (
        viQualityFlagsS.eq(0)
        .And(viQualityFlagsD.eq(0))
        .And(viQualityFlagsS.eq(1))
        .And(viQualityFlagsD.eq(1))
        .And(viSnowIceFlagsD)
        .eq(0)
    )
    # .And(viShadowFlagsD).eq(0); # No shadow
    return image.updateMask(mask)


mod13q1_QC = terra.map(modisQA_mask)
myd13q1_QC = aqua.map(modisQA_mask)

mxd13q1_cleaned = mod13q1_QC.select("EVI").merge(myd13q1_QC.select("EVI"))
mxd13q1_cleaned_sorted = mxd13q1_cleaned.sort("system:time_start")

# Applying ESA WorldCover Cropmask
worldcover = ee.Image('ESA/WorldCover/v200/2021')
crop_mask = worldcover.eq(40).rename("crop").clip(haiti_geometry)
crop_data_viz_dea = crop_mask.mask(crop_mask.gt(0))

# apply cropland mask to imageCollection
def cropmask(img):
    return img.updateMask(crop_mask).clip(haiti_geometry)  # .multiply(scale_factor) # unmask(-1)


mxd13q1 = mxd13q1_cleaned_sorted.map(cropmask)
```

```python
from scipy.signal import savgol_filter

def preprocess_series(series):
    """TIMESAT-style preprocessing."""
    series = series.interpolate(limit_direction="both")  # Step 2: Interpolate
    median = series.rolling(window=5, center=True).median()
    std_dev = series.std()
    series = series.mask((series - median).abs() > 2 * std_dev)  # Step 1: Remove outliers
    series = series.interpolate(limit_direction="both")
    smoothed = savgol_filter(series, window_length=5, polyorder=2)  # Step 3: Smooth
    return smoothed

def extract_sos_mos_eos(smoothed, dates, threshold=0.2):
    max_val = smoothed.max()
    min_val = smoothed.min()
    amp = max_val - min_val
    sos = mos = eos = None
    for i in range(1, len(smoothed)):
        if sos is None and smoothed[i] > min_val + threshold * amp:
            sos = dates[i]
        if smoothed[i] == max_val:
            mos = dates[i]
        if sos is not None and smoothed[i] < min_val + threshold * amp:
            eos = dates[i]
            break
    return sos, mos, eos
```

### Crop Seasonality
Using this time series dataset of EVI images, we apply several pre-processing steps to extract critical phenological parameters: start of season (SOS), middle of season (MOS), end of season (EOS), length of season (LOS), etc. This workflow is heavily inspired by the [TIMESAT](https://web.nateko.lu.se/timesat/timesat.asp) software, although in this implementation we use the [Phenolopy](https://github.com/frontiersi/phenolopy/tree/main) open-source package.  

**Pre-processing steps**  
1. Remove outliers from dataset on per-pixel basis using median method: outlier if median from a moving window < or > standard deviation of time-series times 2.
2. Interpolate missing values linearly
3. Smooth data on per-pixel basis (using Savitsky Golay filter, window length of 3, and polyorder of 1)  

**Phenology Process**  
We then extract crop seasonality metrics using the seasonal amplitude method from the phenolopy package. 

The chart below shows the result of this process for a single crop pixel. The green dots represent the raw EVI values, the black line represents the processed EVI values, and the red dotted lines represent season parameters extracted for that pixel: start of season, peak of season, and end of season.

### Anomalies in EVI

Anomalies are calculated using z score values. 

$$z = \frac{x - \mu}{\sigma}$$

Where:
- $z$ is the z-score (standard score)
- $x$ is the individual median EVI value
- $\mu$ is the mean of the median EVI values
- $\sigma$ is the standard deviation of the median EVI values

The higher the variation the greater the anomaly i.e., if the value is -1 then the EVI for that time period is far lesser than the normal and vice versa. 

The Z scores are calculated for 2023 and 2024 compared to a historic 10 year average. They are then displayed as a histogram, a map and an aggregated map. 

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import LinearSegmentedColormap

def plot_zscore_histograms(zscore_2023_path, zscore_2024_path, output_path=None):
    """
    Create histograms from EVI z-score TIF files for 2023 and 2024.
    
    Parameters:
    zscore_2023_path: Path to 2023 z-score TIF file
    zscore_2024_path: Path to 2024 z-score TIF file
    output_path: Path to save the output figure (optional)
    """
    # Read the raster data
    with rasterio.open(zscore_2023_path) as src_2023:
        zscore_2023 = src_2023.read(1)
        nodata_2023 = src_2023.nodata
    
    with rasterio.open(zscore_2024_path) as src_2024:
        zscore_2024 = src_2024.read(1)
        nodata_2024 = src_2024.nodata
    
    # Filter out nodata values
    if nodata_2023 is not None:
        zscore_2023 = zscore_2023[zscore_2023 != nodata_2023]
    if nodata_2024 is not None:
        zscore_2024 = zscore_2024[zscore_2024 != nodata_2024]
    
    # Filter out extreme outliers (optional)
    # This keeps values between -5 and 5 standard deviations
    zscore_2023 = zscore_2023[(zscore_2023 >= -5) & (zscore_2023 <= 5)]
    zscore_2024 = zscore_2024[(zscore_2024 >= -5) & (zscore_2024 <= 5)]
    
    # Calculate key statistics
    stats_2023 = {
        'mean': np.mean(zscore_2023),
        'median': np.median(zscore_2023),
        'std': np.std(zscore_2023),
        'significant_negative': np.sum(zscore_2023 < -1.96) / len(zscore_2023) * 100,
        'significant_positive': np.sum(zscore_2023 > 1.96) / len(zscore_2023) * 100
    }
    
    stats_2024 = {
        'mean': np.mean(zscore_2024),
        'median': np.median(zscore_2024),
        'std': np.std(zscore_2024),
        'significant_negative': np.sum(zscore_2024 < -1.96) / len(zscore_2024) * 100,
        'significant_positive': np.sum(zscore_2024 > 1.96) / len(zscore_2024) * 100
    }
    
    # Create a figure with 3 subplots
    fig, axes = plt.subplots(1, 1, figsize=(12, 4))
    
    # Custom color palettes
    custom_blue = LinearSegmentedColormap.from_list("", ["lightblue", "darkblue"])
    custom_orange = LinearSegmentedColormap.from_list("", ["peachpuff", "darkorange"])
    
    # # Plot 2023 histogram
    # sns.histplot(zscore_2023, kde=True, bins=50, color='blue', ax=axes[0], stat='density')
    # axes[0].set_title('EVI Z-Score Distribution - 2023 Growing Season', fontsize=14)
    # axes[0].axvline(x=0, color='black', linestyle='--', alpha=0.7)
    # axes[0].axvline(x=-1.96, color='red', linestyle='--', alpha=0.7)
    # axes[0].axvline(x=1.96, color='green', linestyle='--', alpha=0.7)
    # axes[0].set_xlabel('Z-Score', fontsize=12)
    # axes[0].set_ylabel('Density', fontsize=12)
    
    # # Add annotations for 2023
    # axes[0].text(0.05, 0.95, f"Mean: {stats_2023['mean']:.2f}\nMedian: {stats_2023['median']:.2f}\n"
    #             f"Significant Negative: {stats_2023['significant_negative']:.1f}%\n"
    #             f"Significant Positive: {stats_2023['significant_positive']:.1f}%",
    #             transform=axes[0].transAxes, fontsize=10, va='top',
    #             bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    # # Plot 2024 histogram
    # sns.histplot(zscore_2024, kde=True, bins=50, color='orange', ax=axes[1], stat='density')
    # axes[1].set_title('EVI Z-Score Distribution - 2024 Growing Season', fontsize=14)
    # axes[1].axvline(x=0, color='black', linestyle='--', alpha=0.7)
    # axes[1].axvline(x=-1.96, color='red', linestyle='--', alpha=0.7)
    # axes[1].axvline(x=1.96, color='green', linestyle='--', alpha=0.7)
    # axes[1].set_xlabel('Z-Score', fontsize=12)
    # axes[1].set_ylabel('Density', fontsize=12)
    
    # # Add annotations for 2024
    # axes[1].text(0.05, 0.95, f"Mean: {stats_2024['mean']:.2f}\nMedian: {stats_2024['median']:.2f}\n"
    #             f"Significant Negative: {stats_2024['significant_negative']:.1f}%\n"
    #             f"Significant Positive: {stats_2024['significant_positive']:.1f}%",
    #             transform=axes[1].transAxes, fontsize=10, va='top',
    #             bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    # Plot both histograms together for comparison
    sns.histplot(zscore_2023, kde=True, bins=50, color='blue', alpha=0.5, 
                label='2023', ax=axes, stat='density')
    sns.histplot(zscore_2024, kde=True, bins=50, color='orange', alpha=0.5, 
                label='2024', ax=axes, stat='density')
    axes.set_title('Comparison of EVI Z-Score Distributions', fontsize=14)
    axes.axvline(x=0, color='black', linestyle='--', alpha=0.7)
    axes.axvline(x=-1.96, color='red', linestyle='--', alpha=0.7, label='Significantly Below Normal')
    axes.axvline(x=1.96, color='green', linestyle='--', alpha=0.7, label='Significantly Above Normal')
    axes.set_xlabel('Z-Score', fontsize=12)
    axes.set_ylabel('Density', fontsize=12)
    axes.legend(loc='upper right')
    
    # Add interpretation text
    if stats_2023['mean'] < 0 and stats_2024['mean'] < 0:
        comparison = "Both years show below-average vegetation conditions, "
        if stats_2023['mean'] < stats_2024['mean']:
            comparison += "with 2023 being more severely affected than 2024."
        else:
            comparison += "with 2024 being more severely affected than 2023."
    elif stats_2023['mean'] > 0 and stats_2024['mean'] > 0:
        comparison = "Both years show above-average vegetation conditions, "
        if stats_2023['mean'] > stats_2024['mean']:
            comparison += "with 2023 showing more robust growth than 2024."
        else:
            comparison += "with 2024 showing more robust growth than 2023."
    else:
        if stats_2023['mean'] > 0:
            comparison = "2023 showed above-average vegetation conditions, while 2024 showed below-average conditions."
        else:
            comparison = "2023 showed below-average vegetation conditions, while 2024 showed above-average conditions."
    
    axes.text(0.5, 0.05, comparison,
                transform=axes.transAxes, fontsize=10, ha='center',
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    
    # Save the figure if output path is provided
    if output_path:
        plt.savefig(output_path, dpi=300, bbox_inches='tight')
    
    plt.show()

# Example usage:
# plot_zscore_histograms('evi_zscore_2023.tif', 'evi_zscore_2024.tif', 'evi_zscore_histograms.png')
```

```python
import folium
import numpy as np
import rasterio
from rasterio import warp
import branca.colormap as bcm
from matplotlib import colors
import os
from os.path import join, expanduser

# Define the colormap and data range
vmin = -2
vmax = 2
colormap = bcm.linear.BrBG_11.scale(vmin, vmax)

# Set the data directory - this would need to be adjusted to your actual data location
data_dir = '../../docs/images/agriculture/'

def mapvalue2color(value, cmap):
    """
    Map a pixel value of image to a color in the rgba format.
    As a special case, nans will be mapped totally transparent.

    Inputs
        -- value - pixel value of image, could be np.nan
        -- cmap - a linear colormap from branca.colormap.linear
    Output
        -- a color value in the rgba format (r, g, b, a)
    """
    if np.isnan(value):
        return (1, 0, 0, 0)  # Fully transparent for NaN values
    else:
        return colors.to_rgba(cmap(value), 1)  # Full opacity for valid values

# Define map center coordinates (Haiti)
centx = centx 
centy = centy  

# Create a Folium map with dark theme
m = folium.Map(
    location=[centy, centx],
    tiles="CartoDB.DarkMatter",  # Using a dark theme tile
    zoom_start=10
)

# Destination coordinate reference system (CRS) - WGS 84
dest_crs = "EPSG:4326"

# Add each year's EVI anomaly data as a separate layer
for year in range(2023, 2025):
    # Construct the file path - adjust the filename pattern to match your data
    file_path = join(data_dir, f"evi_zscore_{year}.tif")
    
    # Check if the file exists before trying to open it
    if os.path.exists(file_path):
        # Open the GeoTIFF file
        src = rasterio.open(file_path)
        
        # Read the data
        data = src.read(1)
        
        # Get the bounds of the raster
        left, bottom, right, top = src.bounds
        
        # Transform the bounds to the destination CRS (WGS 84)
        left, bottom, right, top = warp.transform_bounds(
            src_crs=src.crs,
            dst_crs=dest_crs,
            left=left,
            bottom=bottom,
            right=right,
            top=top,
        )
        
        # Define the bounds for the image overlay
        bounds = [[bottom, left], [top, right]]
        
        # Set the latest year to be visible by default, hide others
        show = True if year == 2023 else False
        
        # Add the image as an overlay
        folium.raster_layers.ImageOverlay(
            image=data,
            name=f"{year}",
            opacity=1,
            bounds=bounds,
            interactive=False,
            cross_origin=False,
            zindex=1,
            colormap=lambda value: mapvalue2color(value, colormap),
            show=show,
        ).add_to(m)
        
        # Close the raster file
        src.close()

# Add the colormap legend to the map with custom styling for dark background
colormap.caption = "EVI Anomaly, Z-Score"
colormap_html = colormap._repr_html_()
# Modify the legend style to better fit with dark background
styled_colormap = folium.Element(
    colormap_html.replace(
        'background: white',
        'background: #121212; color: white;'
    ).replace(
        'color: #333333',
        'color: #ffffff'
    )
)
m.get_root().html.add_child(styled_colormap)

# Add a layer control to toggle between years
folium.LayerControl().add_to(m)

# Add custom CSS to ensure text appears properly on dark background
custom_css = """
<style>
.leaflet-control-layers,
.leaflet-control-layers-list,
.leaflet-popup-content-wrapper,
.leaflet-popup-tip {
    background-color: #121212 !important;
    color: white !important;
}
.leaflet-control-layers-selector + span {
    color: white !important;
}
.leaflet-bar a {
    background-color: #121212 !important;
    color: white !important;
    border-color: #333 !important;
}
.leaflet-bar a:hover {
    background-color: #333 !important;
}
</style>
"""
m.get_root().html.add_child(folium.Element(custom_css))

# Display the map
m

# Uncomment to save the map to an HTML file
# map_path = './charts/interactive-map-anomalies.html'
# m.save(map_path)
```

```python
import geopandas as gpd
import numpy as np
from rasterstats import zonal_stats

def aggregate_zscores(raster_path, adm1):
    """Aggregate pixel Z-scores to ADM1 regions"""
    # Load admin boundaries
    #adm1 = gpd.read_file(adm1_path)
    
    # Calculate zonal statistics
    stats = zonal_stats(
        adm1,
        raster_path,
        stats=['mean', 'median', 'min', 'max', 'std', 'count'],
        geojson_out=True
    )
    
    # Convert to GeoDataFrame
    result = gpd.GeoDataFrame.from_features(stats)
    
    # Add original attributes
    for col in ['shapeName', 'shapeISO', 'shapeID', 'shapeGroup', 'shapeType']:
        if col in adm1.columns:
            result[col] = adm1[col].values
    
    return result

def plot_zscore_map(gdf, stat='mean', title='EVI Z-Score by Region'):
    """Plot aggregated Z-scores by region"""
    import matplotlib.pyplot as plt
    
    fig, ax = plt.subplots(figsize=(8, 4))
    gdf.plot(column=stat, cmap='BrBG', linewidth=0.5, 
             edgecolor='0.5', legend=True, ax=ax, vmin=-3, vmax=3)
    ax.set_title(title)
    ax.set_axis_off()
    return fig
```

```python
import glob


def process_evi_files(input_dir, output_dir, evi_files):
    
    # List to hold all transformed dataframes
    all_transformed_data = []

    for file_path in evi_files:
        print(f"Processing {os.path.basename(file_path)}...")
        
        try:
            # Read the CSV file
            df = pd.read_csv(file_path)
            
            # Create an empty list to hold the transformed rows
            transformed_data = []
            
            # Identify EVI columns (those ending with _EVI)
            evi_columns = [col for col in df.columns if col.endswith('_EVI')]
            
            # Identify shape/geography columns
            shape_columns = [col for col in df.columns if col.startswith('shape') or col == 'system:index']
            
            # For each row in the original dataframe
            for _, row in df.iterrows():
                admin_area = row['shapeName'] if 'shapeName' in df.columns else row['system:index']
                
                # For each EVI column, create a new row with date information
                for evi_col in evi_columns:
                    # Extract date components from column name
                    # Format is typically like "1_2012_06_09_EVI"
                    try:
                        parts = evi_col.split('_')
                        collection = parts[0]
                        date_year = parts[1]
                        date_month = parts[2]
                        date_day = parts[3]
                        
                        # Create a date string
                        date_str = f"{date_year}-{date_month}-{date_day}"
                        
                        # Create a row for the transformed data
                        new_row = {
                            'date': date_str,
                            'admin_area': admin_area,
                            'collection': collection,
                            'evi_value': row[evi_col]
                        }
                        
                        # Add all shape columns (except '.geo' which can cause issues)
                        for shape_col in shape_columns:
                            if shape_col in row and shape_col != '.geo':
                                new_row[shape_col] = row[shape_col]
                        
                        transformed_data.append(new_row)
                    except (IndexError, ValueError) as e:
                        print(f"Error processing column {evi_col}: {e}")
            
            # Create a dataframe from the transformed data
            if transformed_data:
                transformed_df = pd.DataFrame(transformed_data)
                all_transformed_data.append(transformed_df)
            
        except Exception as e:
            print(f"Error processing {file_path}: {e}")

    # Combine all transformed dataframes
    if all_transformed_data:
        combined_df = pd.concat(all_transformed_data, ignore_index=True)
        
        # Convert date to datetime for proper sorting
        combined_df['date'] = pd.to_datetime(combined_df['date'])
        
        # Sort by admin area and date
        combined_df = combined_df.sort_values(['admin_area', 'date'])
        
        # Save the combined data to a single file
        combined_file = os.path.join(output_dir, "haiti_evi_all_years_combined.csv")
        combined_df.to_csv(combined_file, index=False)
        print(f"Saved combined data to {combined_file}")
        
        # Optional: You can also create monthly and yearly summaries
        monthly_summary = combined_df.groupby(['admin_area', pd.Grouper(key='date', freq='MS')])['evi_value'].agg(['mean', 'min', 'max', 'count']).reset_index()
        monthly_file = os.path.join(output_dir, "haiti_evi_monthly_summary.csv")
        monthly_summary.to_csv(monthly_file, index=False)
        print(f"Saved monthly summary to {monthly_file}")
        
        yearly_summary = combined_df.groupby(['admin_area', pd.Grouper(key='date', freq='YS')])['evi_value'].agg(['mean', 'min', 'max', 'count']).reset_index()
        yearly_file = os.path.join(output_dir, "haiti_evi_yearly_summary.csv")
        yearly_summary.to_csv(yearly_file, index=False)
        print(f"Saved yearly summary to {yearly_file}")
    else:
        print("No data was processed successfully")

    print("Processing complete!")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from scipy import stats
import ipywidgets as widgets
from IPython.display import display

# Function to add trend line to a subplot
def add_trend_line(ax, x_data, y_data, color='grey'):
    """
    Add a linear trend line to the given axis.
    
    Parameters:
    - ax: matplotlib axis to plot on
    - x_data: datetime array for x-axis
    - y_data: EVI values array for y-axis
    - color: color of the trend line
    
    Returns:
    - line: the plotted line object
    """
    # Convert dates to numbers for regression
    x_numeric = mdates.date2num(x_data)
    
    # Perform linear regression
    slope, intercept, r_value, p_value, std_err = stats.linregress(x_numeric, y_data)
    
    # Calculate slope per year (instead of per day)
    slope_per_year = slope * 365.25
    
    # Create trend line
    trend_x = np.array([x_numeric.min(), x_numeric.max()])
    trend_y = intercept + slope * trend_x
    
    # Convert x back to dates for plotting
    trend_x_dates = mdates.num2date(trend_x)
    
    # Plot trend line
    line = ax.plot(trend_x_dates, trend_y, color=color, linestyle='--', linewidth=1.5, )
    
    # Add legend with information about the trend
    ax.legend(loc='upper left', fontsize=8)
    
    return line

# Enhanced plotting function with trend line option
def get_trend_plot(evi_median, crop_admin_regions, title, show_trend=False):
    """
    Create EVI time series plots by admin area with optional trend lines.
    
    Parameters:
    - evi_median: DataFrame with columns 'admin_area', 'date', and 'median'
    - crop_admin_regions: List of admin areas to include in the plot
    - title: Plot title
    - show_trend: Boolean indicating whether to show trend lines
    
    Returns:
    - fig: The matplotlib figure object
    """
    # Convert date to datetime
    evi_median['date'] = pd.to_datetime(evi_median['date'])

    # Get unique admin areas
    admin_areas = crop_admin_regions
    n_areas = len(admin_areas)

    # Create figure and subplots
    fig, axes = plt.subplots(n_areas, 1, figsize=(12, 2*n_areas), sharex=True)
    fig.suptitle(title, fontsize=14)

    # Make axes iterable even when there's only one subplot
    if n_areas == 1:
        axes = [axes]

    # Add EVI label on the left side
    fig.text(0.04, 0.5, 'EVI', va='center', rotation='vertical', fontsize=12)

    # Plot each admin area in its own subplot
    for i, area in enumerate(admin_areas):
        ax = axes[i]
        
        # Filter data for this admin area and sort by date
        area_data = evi_median[evi_median['admin_area'] == area].sort_values('date')
        
        # Plot the data
        ax.plot(area_data['date'], area_data['median'], color='green', linewidth=1.5)
        
        # Add trend line if requested
        if show_trend and len(area_data) > 1:
            add_trend_line(ax, area_data['date'].values, area_data['median'].values)
        
        # Set y-axis limits based on your reference image
        ax.set_ylim(0.20, 0.55)
        ax.set_yticks([0.20, 0.25, 0.30, 0.35, 0.40, 0.45, 0.50, 0.55])
        
        # Add area name as title in the center of the subplot
        ax.set_title(area, fontsize=10, loc='center')
        
        # Add gridlines
        ax.grid(True, linestyle='-', linewidth=0.5, alpha=0.5)
        
        # Remove spines and ticks for cleaner look
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
        ax.spines['bottom'].set_visible(False)
        
        # Only display y-axis on the left
        ax.yaxis.set_ticks_position('left')

    # Format x-axis to show years
    axes[-1].xaxis.set_major_locator(mdates.YearLocator())
    axes[-1].xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    axes[-1].tick_params(axis='x', rotation=45)

    # Add source note
    fig.text(0.95, 0.01, 'Source: MODIS Processed', ha='right', fontsize=8)

    # Adjust layout
    plt.tight_layout()
    plt.subplots_adjust(hspace=0.5, left=0.1, bottom=0.1, top=0.95)

    return fig

# Function to create an interactive plot with a toggle
def interactive_evi_plot(evi_median, crop_admin_regions):
    """
    Create an interactive EVI plot with a checkbox to toggle trend lines.
    
    Parameters:
    - evi_median: DataFrame with columns 'admin_area', 'date', and 'median'
    - crop_admin_regions: List of admin areas to include in the plot
    """
    # Define the update function for the checkbox
    def update_plot(show_trend):
        plt.close('all')  # Close previous plots
        title = 'EVI Time Series by Admin Area'
        fig = get_trend_plot(evi_median, crop_admin_regions, title, show_trend)
        plt.show()
    
    # Create the checkbox widget
    checkbox = widgets.Checkbox(
        value=False,
        description='Show Trend Line',
        disabled=False,
        indent=False
    )
    
    # Set up the interactive widget
    widgets.interactive(update_plot, show_trend=checkbox)
    
    # Show the initial plot
    update_plot(False)
    
    # Display the checkbox
    display(checkbox)

# Example usage in a Jupyter notebook:
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def plot_evi_comparison(df, reference_df, window_size=3):
    # Make sure we have date as datetime
    # df['date'] = pd.to_datetime(df['date'])
    # df['month'] = df['date'].dt.month
    
    # Filter for 2023 and 2024 data
    df_2023 = df[df['year'] == 2023]
    df_2024 = df[df['year'] == 2024]
    
    # Group data by admin_area and month
    data_2023 = df_2023.groupby(['admin_area', 'month'])['median'].mean().reset_index()
    data_2024 = df_2024.groupby(['admin_area', 'month'])['median'].mean().reset_index()
    
    # Get the reference data
    ref_data = reference_df.groupby(['admin_area', 'month'])['median_ref'].mean().reset_index()
    
    # Plot for each admin area
    admin_areas = sorted(df['admin_area'].unique())
    num_areas = len(admin_areas)
    
    # Create a figure with appropriate layout based on number of admin areas
    if num_areas == 1:
        fig, ax = plt.subplots(figsize=(12, 4))
        axes = [ax]  # Make a list with the single axis for consistent handling
    else:
        fig, axes = plt.subplots(num_areas, 1, figsize=(12, 2 * num_areas), sharex=True)
        if num_areas == 1:  # In case plt.subplots returns a single axis instead of array
            axes = [axes]
    
    for i, area in enumerate(admin_areas):
        area_ref = ref_data[ref_data['admin_area'] == area]
        area_2023 = data_2023[data_2023['admin_area'] == area]
        area_2024 = data_2024[data_2024['admin_area'] == area]
        
        # Apply smoothing for each dataset (if enough data points)
        if len(area_ref) >= window_size:
            # For monthly data, we need to handle the circular nature of months
            # Create extended series to handle wrap-around for smoothing
            months_ext = np.concatenate([area_ref['month'].values - 12, 
                                         area_ref['month'].values, 
                                         area_ref['month'].values + 12])
            values_ext = np.concatenate([area_ref['median_ref'].values, 
                                         area_ref['median_ref'].values, 
                                         area_ref['median_ref'].values])
            
            # Create a DataFrame with extended series
            ext_df = pd.DataFrame({'month': months_ext, 'value': values_ext})
            ext_df = ext_df.sort_values('month')
            
            # Apply rolling average
            ext_df['smoothed'] = ext_df['value'].rolling(window=window_size, center=True).mean()
            
            # Extract the middle part (original range)
            smoothed_ref = ext_df[(ext_df['month'] >= 1) & (ext_df['month'] <= 12)]
            
            # Plot smoothed reference line in grey
            axes[i].plot(smoothed_ref['month'], smoothed_ref['smoothed'], 'grey', 
                        linestyle='--', label='Historical Ref (Smoothed)')
        else:
            # Not enough data points, plot raw data
            axes[i].plot(area_ref['month'], area_ref['median_ref'], 'grey', 
                        linestyle='--', label='Historical Ref')
        
        # Apply similar smoothing for 2023 and 2024 data if enough data points
        if len(area_2023) >= window_size:
            # Apply similar circular month handling for 2023
            months_ext = np.concatenate([area_2023['month'].values - 12, 
                                         area_2023['month'].values, 
                                         area_2023['month'].values + 12])
            values_ext = np.concatenate([area_2023['median'].values, 
                                         area_2023['median'].values, 
                                         area_2023['median'].values])
            
            ext_df = pd.DataFrame({'month': months_ext, 'value': values_ext})
            ext_df = ext_df.sort_values('month')
            ext_df['smoothed'] = ext_df['value'].rolling(window=window_size, center=True).mean()
            smoothed_2023 = ext_df[(ext_df['month'] >= 1) & (ext_df['month'] <= 12)]
            
            # Plot smoothed 2023 data in blue
            axes[i].plot(smoothed_2023['month'], smoothed_2023['smoothed'], 'blue', 
                        label='2023 (Smoothed)')
        elif not area_2023.empty:
            # Not enough data points, plot raw data
            axes[i].plot(area_2023['month'], area_2023['median'], 'blue', label='2023')
        
        # Similar approach for 2024
        if len(area_2024) >= window_size:
            months_ext = np.concatenate([area_2024['month'].values - 12, 
                                         area_2024['month'].values, 
                                         area_2024['month'].values + 12])
            values_ext = np.concatenate([area_2024['median'].values, 
                                         area_2024['median'].values, 
                                         area_2024['median'].values])
            
            ext_df = pd.DataFrame({'month': months_ext, 'value': values_ext})
            ext_df = ext_df.sort_values('month')
            ext_df['smoothed'] = ext_df['value'].rolling(window=window_size, center=True).mean()
            smoothed_2024 = ext_df[(ext_df['month'] >= 1) & (ext_df['month'] <= 12)]
            
            # Plot smoothed 2024 data in orange
            axes[i].plot(smoothed_2024['month'], smoothed_2024['smoothed'], 'orange', 
                        label='2024 (Smoothed)')
        elif not area_2024.empty:
            # Not enough data points, plot raw data
            axes[i].plot(area_2024['month'], area_2024['median'], 'orange', label='2024')
        
        axes[i].set_title(area)
        axes[i].set_ylabel('EVI')
        axes[i].grid(True, linestyle='--', alpha=0.7)
        axes[i].set_xlim(1, 12)      # Set x-limits for months 1-12
        
        # Always show legend on single plots
        if num_areas == 1 or i == 0:
            axes[i].legend()
    
    # Only set xlabel on the bottom plot
    axes[-1].set_xlabel('Month')
    plt.tight_layout()
    plt.show()
```

```python
plot_evi_comparison(evi_median_adm0, evi_median_adm0_ref)
```

**2024 was a better year for crop yield compared to the historic average.** Although 2023 had a lower values in the beginning, the growing season saw greater crop yield.
